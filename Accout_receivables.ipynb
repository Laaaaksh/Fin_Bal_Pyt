```python
import requests
import pandas as pd

acc_Receivables = {}
#creating a dictionary for receiving the account receivables

stock = 'AAPL'
api_key = '08e2d2a138dba992cc772bd27479aade'

bS = requests.get(f'https://financialmodelingprep.com/api/v3/balance-sheet-statement/{stock}?limit=10&apikey={api_key}').json()

print(bS)

acc_Receivables[stock] = {}
#creating a nested dictionary so that we know the particualr company we are addressing too!
for item in bS:
  receivables = item['netReceivables']
  period = item['date']

  acc_Receivables[stock][period] = receivables

print(acc_Receivables)

receivables_apple = pd.DataFrame.from_dict(acc_Receivables)
#creating a dataframe from the dictionary

print(receivables_apple)

import plotly.express as px

fig = px.bar(receivables_apple, x=receivables_apple.index , y=receivables_apple[stock], title = 'Acc Receivables ' + stock)
fig.show()

companies = ['IBM','CSCO','ORCL','AMZN','MSFT','AAPL','GOOG',]

acc_Receivables = {}

for company in companies:
  balanceSheet = requests.get(f'https://financialmodelingprep.com/api/v3/balance-sheet-statement/{company}?limit=1&apikey={api_key}').json()
  acc_Receivables[company] = {}
  receivables = balanceSheet[0]['netReceivables']
  acc_Receivables[company]['receivables'] = receivables

  IS = requests.get(f'https://financialmodelingprep.com/api/v3/income-statement/{company}?limit=1&apikey={api_key}').json()
  revenues = IS[0]['revenue']
  acc_Receivables[company]['revenue'] = revenues
  acc_Receivables[company]['receivable_Turnover'] = revenues / receivables
  acc_Receivables[company]['DSO'] = 365 / acc_Receivables[company]['receivable_Turnover']





receivable_Companies = pd.DataFrame.from_dict(acc_Receivables)
receivable_Companies

receivable_Companies_DSO = receivable_Companies[receivable_Companies.index == 'DSO']

receivable_Companies_DSO = receivable_Companies_DSO.T

fig = px.bar(receivable_Companies_DSO, x=receivable_Companies_DSO.index, y=receivable_Companies_DSO['DSO'])
fig.show()
```
