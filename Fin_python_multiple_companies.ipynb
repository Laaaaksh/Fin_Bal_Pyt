```python
import requests
import pandas as pd
api_key = '08e2d2a138dba992cc772bd27479aade'
screen = requests.get(f'https://financialmodelingprep.com/api/v3/stock-screener?marketCapMoreThan=1000000000&sector=Technology&exchange=NASDAQ&limit=10&apikey={api_key}').json()
#a method to filter out a certain companies on basis of certain parameters , the parameters are seperated by end
#v3 is the version of the api that we are using.
#print(screen)

values = []
for item in screen:
  values.append(item['symbol'])

#print(tickers)
#this will print the list of the companies as the symbol was used to give there name

company_list = {} 
for item in values:
  BS = requests.get(f'https://financialmodelingprep.com/api/v3/balance-sheet-statement/{item}?limit=1&apikey={api_key}').json()
  company_list[item] = BS[0]
  #extracting the first element from the balance sheet!

 
BS_industry = pd.DataFrame.from_dict(company_list)
BS_industry

#if you are not on google collab then you need to use the print statement as well.

BS_our_Columns = ['date', 
       'cashAndCashEquivalents', 'shortTermInvestments',
       'cashAndShortTermInvestments', 'netReceivables', 'inventory',
       'otherCurrentAssets', 'totalCurrentAssets', 'propertyPlantEquipmentNet',
       'goodwill', 'intangibleAssets', 'goodwillAndIntangibleAssets',
       'longTermInvestments', 'taxAssets', 'otherNonCurrentAssets',
       'totalNonCurrentAssets', 'otherAssets', 'totalAssets',
       'accountPayables', 'shortTermDebt', 'taxPayables', 'deferredRevenue',
       'otherCurrentLiabilities', 'totalCurrentLiabilities', 'longTermDebt',
       'deferredRevenueNonCurrent', 'deferredTaxLiabilitiesNonCurrent',
       'otherNonCurrentLiabilities', 'totalNonCurrentLiabilities',
       'otherLiabilities', 'totalLiabilities', 'commonStock',
       'retainedEarnings', 'accumulatedOtherComprehensiveIncomeLoss',
       'othertotalStockholdersEquity', 'totalStockholdersEquity',
       'totalLiabilitiesAndStockholdersEquity', 'totalInvestments',
       'totalDebt', 'netDebt']

       #removing the unnecssary columns

BS_industry = BS_industry.T
BS_industry = BS_industry[BS_our_Columns]
BS_industry = BS_industry.T
#print(BS_industry)

BS_industry1 = BS_industry.iloc[1:]/1000000
#so that we get small values
BS_industry1 = BS_industry1.T
BS_industry1['cashAndCashEquivalents'].plot(legend=True, figsize=(15,5))
#plotting a particular column form the transpose table.


BS_industry1.T
```
